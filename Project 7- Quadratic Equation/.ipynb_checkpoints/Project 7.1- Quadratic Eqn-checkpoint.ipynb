```python
import cmath
def Quadratic (a,b,c):
    e,f = [str(x) if x < 0 else '+ ' +str(x) for x in [b,c]]
    tom = str(a) +'x**2 '+ str(e) + 'x ' + str(f) if a != 1 else 'x**2 '+ str(e) + 'x ' + str(f)
    print('The equation you wish to solve is: ' + str(tom)+ ' = 0')
    d = b**2 - 4*a*c
    soln1 = (-b + (cmath.sqrt(d)).real)/(2*a)
    soln2 = (-b - (cmath.sqrt(d)).real)/(2*a)
    eqn1 = '(x + ' +str(-soln1)+ ')' if (-soln1) > 0 else  '(x - ' +str(soln1)+ ')'
    eqn2 = '(x + ' +str(-soln2)+ ')' if (-soln2) > 0 else  '(x - ' +str(soln2)+ ')'
    return 'Therefore, ' +str(tom)+ ' = ' +str(eqn1)+str(eqn2)
```

```python
Quadratic(2,3,4)
```
