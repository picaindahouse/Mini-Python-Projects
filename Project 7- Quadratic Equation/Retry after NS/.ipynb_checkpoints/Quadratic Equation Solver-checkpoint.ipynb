```python
from math import sqrt
def solution(a,b,c):
    if (b**2 - 4*a*c) < 0:
        return 0
    soln1 = round((-b + sqrt(b**2 - 4*a*c))/(2*a),3)
    soln2 = round((-b - sqrt(b**2 - 4*a*c))/(2*a),3)
    return (soln1,soln2)
```
